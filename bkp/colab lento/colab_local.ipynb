```python
# prompt: abrir pasta do google drive e substituir essa linha:
# model = torch.hub.load('ultralytics/yolov5', 'custom', "best.pt", force_reload=True)

from flask import Flask, request, jsonify, send_file
import torch
import cv2
import numpy as np
from io import BytesIO
from google.colab import drive
from pyngrok import ngrok

# Autentica o ngrok com seu token
ngrok.set_auth_token("2lfCLIiDqWvSM4R0YQX2vdUN6SM_4RAD6PCVCUKcKBs1bMBpW")

public_url = ngrok.connect(5000)
print("Public URL:", public_url)

drive.mount('/content/drive')

app = Flask(__name__)

# Carrega o modelo YOLOv5
model = torch.hub.load('ultralytics/yolov5', 'custom', "/content/drive/MyDrive/IC 2024/best.pt", force_reload=True)

def detect_bounding_box(frame, conf_threshold=0.8):
    results = model(frame)
    detections = results.pred[0]
    for *box, conf, cls in detections:
        if conf >= conf_threshold:
            x1, y1, x2, y2 = map(int, box)
            label = f'{results.names[int(cls)]} {conf:.2f}'
            color = (0, 255, 0)  # Cor do ret√¢ngulo
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    return frame

@app.route('/process_image', methods=['POST'])
def process_image():
    file = request.files['image']
    img_bytes = file.read()

    nparr = np.frombuffer(img_bytes, np.uint8)
    img_np = cv2.imdecode(nparr, cv2.IMREAD_COLOR)

    # Processa a imagem
    processed_image = detect_bounding_box(img_np)

    # Codifica a imagem de volta para JPEG
    _, buffer = cv2.imencode('.jpg', processed_image)
    img_io = BytesIO(buffer)

    return send_file(img_io, mimetype='image/jpeg')

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
